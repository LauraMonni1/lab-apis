```python
# Your answer here
#1. List Your Favorite Artists

artists = ["Manu Chao", "Fabrizio De André", "Placebo"]
```

```python
#2. Fetch Top Tracks

def get_top_tracks(artist:str):
    
    results = sp.search(q = artist, limit=50)  #search for the artist (function argument)
    tracks = results["tracks"]["items"]        # "actual tracks" in "tracks" dictionary
    popular_tracks = sorted(tracks, key = lambda x: x["popularity"], reverse = True)  # sorting the tracks according to their popularity
    
    #top_5 = [tracks["name"] for tracks in popular_tracks[:5]]  # iterating in the popular_tracks dictionary to "export" the first 5 as list
    # This list comprehension did not handle duplicates (that I found in "Placebo" list) so I did a for loop 
    #to check for duplicates by first creating a set of unique names
    
    top_5 = []
    unique_tracks = set()
    
    for tracks in popular_tracks:
        if len(top_5) == 5:
            break
        if tracks["name"] not in unique_tracks:
            top_5.append(tracks["name"])
            unique_tracks.add(tracks["name"])
    
    return top_5
```

```python
get_top_tracks("Manu Chao")
```

```python
get_top_tracks("Fabrizio De André")
```

```python
get_top_tracks("Placebo")
```

```python
# 3.Discover Related Artists
#this function should accept an artist's name and 
#return the names of the first 5 artists related to the provided artist.
#Store the results in a list named related_artists_list.

def find_related_artists (artist:str):
    results = sp.search(q = artist, limit = 1)
    artist_id = results["tracks"]["items"][0]["artists"][0]["id"]
    related = sp.artist_related_artists(artist_id)
    
    related_names = [artist["name"] for artist in related["artists"]]
    
    return related_names
```

```python
find_related_artists("Manu Chao")
```

```python
relations = [find_related_artists(artist) for artist in artists]
relations
```

## Playlists

The `sp.featured_playlists()` method in `spotipy` fetches a list of Spotify's featured playlists at a given moment. These are curated playlists that Spotify often highlights on the platform's homepage. The method provides a snapshot of the playlists that are being promoted or featured by Spotify at the time of the request.

Once you've fetched the featured playlists, you can extract their IDs (and other details).

```python
playlist_tot = sp.featured_playlists() # We get a playlist id of a playlist we like
playlist_tot
```

```python
playlist_tot["playlists"]["items"][0]["id"]
```

### Getting a Playlist's Details
To fetch details about a specific playlist, you can use the playlist method. You'll need the playlist's Spotify ID.

```python
playlist_id = "37i9dQZF1DXd9zR7tdziuQ"
playlist = sp.playlist(playlist_id)
```

```python
print(playlist['name'])  # Print the playlist's name
print(playlist['description'])  # Print the playlist's description
```

### Getting Tracks from a Playlist
If you want to get the tracks from a specific playlist, you can use the playlist_tracks method.

```python
tracks = sp.playlist_tracks(playlist_id)
for track in tracks['items']:
    print(track['track']['name'])  # Print each track's name
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

```python
def get_top_10(playlist_id):
    
    """The function returns the top 10 tracks from a playlist given its id."""
    
    tracks = sp.playlist_tracks(playlist_id)
    
    top_10 = []
    
    for track in tracks['items'][:10]:
        top_10.append(track['track']['name'])
    
    return top_10
```
